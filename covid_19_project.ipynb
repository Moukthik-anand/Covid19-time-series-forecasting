# Model Building

```python
# Installing prophet
# pip install prophet
'''
Prophet is a time series forecasting package developed by Facebook. It is a efficient for wide range of time series data.
It's particularly useful for data that has strong seasonal effects and when forecasts are needed.
It includes data or events that can impact the forecast.
'''
```

```python
# The columns names should be ds and y and its mandatory
confirmed_cases = df.groupby(['Date'])['Confirmed'].sum().reset_index()
confirmed_cases
```

```python
confirmed_cases.columns = ['ds','y']
confirmed_cases
```

```python
from prophet import Prophet
```

```python
model = Prophet()
```

```python
model.fit(confirmed_cases)
```

```python
future = model.make_future_dataframe(periods=21)
forcasted = model.predict(future)
forcasted
```

```python
'''
ds : [-15,10,25]
This column contains the dates for which the predications have been made.
it stands for "date stamp"

yhat:
This column contains the predicted values for the target variable.
It represents models best estimate for that point in time.

yhat_upper:
This column represents the upper bound of the predicted value.

yhat_lower:
This column represents the lower bound of the predicted value.
```

```python
forcasted[['yhat_lower','yhat','yhat_upper']]
```

```python
model.plot(forcasted)
plt.show()
```
